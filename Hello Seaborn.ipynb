```python
import numpy as np
from datetime import datetime, timedelta


np.random.seed(42)


start_date = datetime(2017, 1, 6)
date_list = [start_date + timedelta(days=x) for x in range(100)]


data = {
    'Shape of You': np.linspace(1, 20000000, 100) ,
    'Despacito': np.sin(np.linspace(0, 4 * np.pi, 100)) * 5000000 ,
    'Something Just Like This': np.exp(np.linspace(0, 5, 100)) * 500000 ,
    'HUMBLE.': -np.linspace(1, 20000000, 100),
    'Unforgettable': np.cos(np.linspace(0, 3 * np.pi, 100)) * 2000000 
}


df = pd.DataFrame(data, index=date_list)


df
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

Months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

data = {

    'Revenue': np.random.randint(10000, 50000, 12),
    'Expenses': np.random.randint(5000, 30000, 12),
    'Profit': np.random.randint(2000, 20000, 12)
}

df_songs = pd.DataFrame(data,index=Months)

df_songs
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=df_songs,annot=True)
```

```python
import numpy as np
import random


random.seed(42)


data = {
    'age': [random.randint(18, 65) for _ in range(1000)],
    'sex': [random.choice(['male', 'female']) for _ in range(1000)],
    'bmi': [round(random.uniform(18.5, 35.0), 2) for _ in range(1000)],
    'children': [random.randint(0, 5) for _ in range(1000)],
    'smoker': [random.choice(['yes', 'no']) for _ in range(1000)],
    'region': [random.choice(['southwest', 'southeast', 'northwest', 'northeast']) for _ in range(1000)],
    'charges': [round(random.uniform(1000, 30000), 2) for _ in range(1000)],
}

df_insurance = pd.DataFrame(data)
df_insurance
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(y=df_insurance['bmi'],x=df_insurance['charges'])
```

```python
plt.figure(figsize=(10,6))
sns.regplot(y=df_insurance['bmi'],x=df_insurance['charges'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(y=df_insurance['bmi'],x=df_insurance['charges'],hue=df_insurance['smoker'])
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(x='bmi',y='charges',hue='smoker',data=df_insurance)
```

```python
sns.swarmplot(x=df_insurance['smoker'],y=df_insurance['charges'])
```

**kernel density estimate (KDE)**

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'],fill=True)
```

```python
#2D KDE plots
sns.jointplot(x=iris_data['Petal Length (cm)'],y=iris_data['Sepal Width (cm)'],kind='kde')
```

**Color-coded plots**

data= provides the name of the variable that we used to read in the data
x= sets the name of column with the data we want to plot
hue= sets the column we'll use to split the data into different histograms

```python
sns.histplot(data=iris_data, x='Petal Length (cm)' , hue='Species')
```

```python
#KDE plots for each species
sns.kdeplot(data=iris_data, x='Petal Length (cm)',hue='Species',fill=True)
```
