```python
loan_term_hist = df['Loan_Amount_Term'].dropna()  

# Plotting the histogram
plt.figure(figsize=(10, 6))
sns.histplot(loan_term_hist, bins=30, color='skyblue')
```

```python
#violin plot on coapplicant income
# a hybrid of a box plot and a kernel density plot, which shows peaks in the data
plt.figure(figsize=(12, 8))


sns.violinplot(x='CoapplicantIncome', data=df, color='lightblue')
```

```python

plt.figure(figsize=(8, 6))

gender_counts = df['Gender'].value_counts()

sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='pastel')
```

```python
x=df['ApplicantIncome']
y=df['LoanAmount']
plt.scatter(x,y)
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
```

```python
#count plot use hue=loan_status education
plt.figure(figsize=(10, 6))
sns.countplot(x='Education', hue='Loan_Status', data=df, palette='Set2')
```

```python
#barplot Gender,ApplicantIncome
sns.barplot(x=df['Gender'],y=df['ApplicantIncome'])
```

```python
#box plot self emmployed and applicant income
plt.figure(figsize=(10, 6))
sns.boxplot(x='Self_Employed', y='ApplicantIncome', data=df, palette='Set3')
```

```python
#heatmap
#we can not plot string data types

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(12, 8))
sns.heatmap(data=df[numeric_columns].corr(), annot=True, linewidths=.5)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='ApplicantIncome', hue='Loan_Status', data=df, palette='Set2')
```
